```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import shutil

def read_data(path) :
    folder_path = path
    file_name = ''
    glob_file = glob.glob(os.path.join(folder_path, file_name + '*txt'))

    for i, v in enumerate(glob_file) :
        df = pd.read_csv(v, sep="\s+", header=0)
        values = df.values
        
        df = df.to_numpy()
        df = np.insert(df, 0, values[:,0], axis=1)
        df = np.insert(df, 1, values[:,4], axis=1)

        df = df[:,:2]

        data_file = os.path.split(v)

        np.savetxt(str(i) + '.csv', df, delimiter=", ", fmt='%s')
        
def make_folder(path, file_name, new_path, new_folder):
    dir = path + '/'
    print(os.path.join(dir, file_name))
    glob_file = glob.glob(os.path.join(dir, file_name))
    # print(glob_file)

    new_dir = new_path + '/' + new_folder 

    try:
        if not os.path.exists(new_dir) :
            os.makedirs(new_dir)
    except OSError:
        print('Error')

    for i, path_lst in enumerate(glob_file):
        data_name = os.path.split(path_lst)
        print(data_name[1])
        if os.path.isfile(new_dir+'/' + data_name[1]):
            print("파일이 이미 존재함: " + new_dir + '/' + data_name[1])
        else:
            shutil.move(dir+data_name[1], new_dir)
        
read_data('/Users/leeyumin/Desktop/handcream/sample_2/')

make_folder('.', '*csv', '/Users/leeyumin/Desktop/handcream/', "sample2")
```
